```python
import cv2
import numpy as np
import os
import matplotlib.pyplot as plt

# Тут указать путь до папки, где находятся изображения
folder_path = './data/building3'


image_paths = [os.path.join(folder_path, filename) for filename in os.listdir(folder_path) if filename.endswith(".jpg")]
images = [cv2.imread(path) for path in image_paths]

# используем Stitcher_create для склеивания 
stitcher = cv2.Stitcher_create()  

(status, panorama) = stitcher.stitch(images)

if status == cv2.Stitcher_OK:
    
    plt.imshow(cv2.cvtColor(panorama, cv2.COLOR_BGR2RGB))
    plt.show()
else:
    print(f"Не удалось создать панораму, ошибка кода статуса: {status}")
```
